# ZOMATO DATA ANALYSIS PROJECT

# STEP 1 -> Importing Python Libraries

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

# STEP 2 -> Load csv data

```python
df = pd.read_csv(r'Zomato data .csv')
```

```python
df.head()
```

```python
df.shape
```

# STEP 3 -> Data Cleaning

```python
df.info()
```

**Converting the data type of `rate` col**

```python
# function to convert rate type (object -> float ) ex: 4.5/5 -> 4.5

def handleRate(value):
    value= str(value).split('/')
    value=value[0]
    return float(value)
```

```python

df['rate'] =df['rate'].apply(handleRate)
```

```python
df.info()
```

```python
df.head()
```

```python
df.isna().sum()
```

# Q-1) Type of Restaurant

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel("Type of Restaurant")
```

# conclusion 1 -> majority of restaurant fall in dining category

# Q-2) Votes each type of restaurant received by customer

```python
df.head()
```

```python
groupedData = df.groupby('listed_in(type)')['votes'].sum()
```

```python
groupedData
```

```python
result = pd.DataFrame({'votes' : groupedData})
result
```

```python
plt.plot(result, c='blue', marker='o')
plt.xlabel("Types of Restaurant", c='red', size='15')
plt.ylabel('Votes', c='red', size='15')
```

# Conclusion 2 -> dining restaurant got more votes

# Q-3) ratings majority of restaurant has received

```python
df.head()
```

```python
plt.hist(df['rate'], bins=4)
plt.title('Rating Distributions')
plt.show()
```

# Conclusion 3 -> majority restaurant received rating from 3.5 to 4.0

# Q-4) Average order spending by couples

```python
df.head()
```

```python
couples_Data = df['approx_cost(for two people)']
couples_Data
```

```python
sns.countplot(x=couples_Data)
```

# Conclusion 4 -> majority of customers prefer restaurant with an order of 300 Rs.

# Q-5) Which mode (online/offline) receive maximum rating

```python
df.head()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x ='online_order', y ='rate', data= df)
```

# Conclusion 5 -> Online order receive higher rating in comparison to offline order 

```python
heatData = df.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(heatData, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Heatmap')
```

# Main Conclusion from Heatmap :
**`Dining Restaurant` primarily accept more `offline order`, `cafe` accepts more `online`, 
`buffet` accepts `online` and `others` also `online`.**
